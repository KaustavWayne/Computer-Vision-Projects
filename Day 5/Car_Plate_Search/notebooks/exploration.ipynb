```python
from ultralytics import YOLO

model = YOLO("yolo11m.pt")
```

```python
# Change data.yaml file 
```

```python
# train

model.train(
    data="dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)
```

```python
# Evaluate on test

best_model = YOLO("runs/detect/train/weights/best.pt")
best_model.val(data="dataset/data.yaml", split="test")
```

 split="test" = evaluate on test folder

 default = evaluate on val folder

```python
# Predict and save results (for visualization)

best_model.predict(source="dataset/test/images", save=True, conf=0.4)
```
